```python
import gradio as gr
```

```python
def correct(sentence):
    res = gf.correct(sentence)
    return res
```

```python
correct('I like for walks')
```

```python
app_inputs = gr.inputs.Textbox(lines=2, placeholder="Enter sentences here.....")
```

```python
interfaces = gr.Interface(fn = correct,
                         inputs = app_inputs,
                         outputs = 'text')
```

```python
interfaces.launch(share=True)
```
